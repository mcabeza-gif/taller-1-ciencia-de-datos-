````python
class Cliente:
    def __init__(self, nombre, email, sector):
        if not validar_email(email):
            raise ValueError("Email no válido.")
        self.id = generar_id()
        self.nombre = nombre
        self.email = email
        self.sector = sector
        self._historial_proyectos = []  # Protegido

    def agregar_al_historial(self, proyecto_nombre):
        self._historial_proyectos.append(proyecto_nombre)

    def __str__(self):
        return f"Cliente: {self.nombre} ({self.sector}) - ID: {self.id}"

    def __repr__(self):
        return f"Cliente('{self.nombre}', '{self.email}', '{self.sector}')"
````

````python
class Dataset:
    def __init__(self, nombre, fuente, filas, columnas, estado="crudo"):
        self.id = generar_id()
        self.nombre = nombre
        self.fuente = fuente
        self.fecha_registro = datetime.now().strftime("%Y-%m-%d %H:%M:%S")

        self._filas = None
        self._columnas = None
        self.filas = filas          # Llama al validador de abajo
        self.columnas = columnas    # Llama al validador de abajo

        estados_validos = ["crudo", "limpio", "procesado"]
        self.estado = estado if estado in estados_validos else "crudo"

    @property
    def filas(self):
        return self._filas

    @filas.setter
    def filas(self, valor):
        if valor < 0:
            raise ValueError("❌ Las filas no pueden ser negativas.")
        self._filas = valor

    @property
    def columnas(self):
        return self._columnas

    @columnas.setter
    def columnas(self, valor):
        if valor < 0:
            raise ValueError("❌ Las columnas no pueden ser negativas.")
        self._columnas = valor

    def calidad(self):
        if self.estado == "procesado": return 99.5
        if self.estado == "limpio": return 95.0
        return 70.0

    def __str__(self):
        return f"Dataset: {self.nombre} ({self.estado}) - {self.filas}x{self.columnas}"
````

````python
class Analista:
    MAX_PROYECTOS = 3  # Puedes cambiar este número si el analista puede hacer más

    def __init__(self, nombre, email, especialidad):
        if not validar_email(email):
            raise ValueError("❌ Error: El formato del email del analista no es válido.")
        self.nombre = nombre
        self.email = email
        self.especialidad = especialidad
        self.proyectos_activos = 0

    def disponible(self):
        return self.proyectos_activos < self.MAX_PROYECTOS

    def __str__(self):
        return f"Analista: {self.nombre} ({self.especialidad}) | Activos: {self.proyectos_activos}"
````

````python
class Proyecto:
    def __init__(self, nombre, descripcion, cliente):
        if not isinstance(cliente, Cliente):
            raise TypeError("❌ El cliente debe ser un objeto válido de la clase Cliente.")

        self.nombre = nombre
        self.descripcion = descripcion
        self.cliente = cliente
        self.datasets = []
        self.equipo = []
        self._estado = "planificación"  # Estado inicial por defecto
        self._historial_estados = [f"Creado en 'planificación' el {datetime.now().strftime('%Y-%m-%d %H:%M:%S')}"]

        # Guardar en el historial del objeto cliente automáticamente
        self.cliente.registrar_proyecto_en_historial(self.nombre)

    @property
    def estado(self):
        return self._estado

    def agregar_dataset(self, dataset):
        if not isinstance(dataset, Dataset):
            raise TypeError("❌ Solo se pueden agregar objetos de tipo Dataset.")
        if dataset not in self.datasets:
            self.datasets.append(dataset)
            print(f"🔹 Dataset '{dataset.nombre}' vinculado al proyecto.")

    def asignar_analista(self, analista):
        if not isinstance(analista, Analista):
            raise TypeError("❌ Solo se pueden asignar objetos de tipo Analista.")

        if not analista.disponible():
            print(f"⚠️ No se pudo asignar: {analista.nombre} no tiene disponibilidad (Límite máximo alcanzado).")
            return False

        if analista not in self.equipo:
            self.equipo.append(analista)
            analista.proyectos_activos += 1
            print(f"🏃 Analista '{analista.nombre}' asignado al proyecto.")
            return True
        return False

    def cambiar_estado(self, nuevo_estado):
        estados_validos = ["planificación", "en_progreso", "completado"]
        if nuevo_estado not in estados_validos:
            raise ValueError(f"❌ Estado no válido. Debe ser uno de: {estados_validos}")

        if nuevo_estado != self._estado:
            registro = f"Cambio: {self._estado} -> {nuevo_estado} ({datetime.now().strftime('%Y-%m-%d %H:%M:%S')})"
            self._historial_estados.append(registro)
            self._estado = nuevo_estado
            print(f"🔄 Estado del proyecto actualizado a: {nuevo_estado.upper()}")

    # 5. Generar reportes básicos (Por Proyecto)
    def reporte(self):
        print("\n" + "="*50)
        print(f"📋 REPORTES DE PROYECTO: {self.nombre.upper()}")
        print(f"Estado: {self._estado.upper()}")
        print(f"Descripción: {self.descripcion}")
        print(f"Cliente: {self.cliente.nombre} | Sector: {self.cliente.sector}")
        print("-"*50)
        print(f"Equipo Técnico Asignado ({len(self.equipo)}):")
        for analista in self.equipo:
            print(f"  • {analista.nombre} ({analista.especialidad})")
        print(f"Datasets Utilizados ({len(self.datasets)}):")
        for ds in self.datasets:
            print(f"  • {ds.nombre} | Calidad: {ds.calidad()}%")
        print(f"Historial de Auditoría:")
        for cambio in self._historial_estados:
            print(f"  [Log] {cambio}")
        print("="*50 + "\n")
````

````python
class DataCoSystem:
    def __init__(self):
        self.proyectos = []

    def registrar_proyecto(self, proyecto):
        if isinstance(proyecto, Proyecto):
            self.proyectos.append(proyecto)

    # 4. Buscar y filtrar información
    def filtrar_proyectos_por_estado(self, estado_buscado):
        return [p for p in self.proyectos if p.estado == estado_buscado]

    # 5. Reporte General del Estado del Negocio
    def generar_reporte_sistema(self):
        print("\n📊 === REPORTE GENERAL DE LA EMPRESA (DATACO) ===")
        print(f"Cantidad total de proyectos gestionados: {len(self.proyectos)}")
        print(f"  - Proyectos en Planificación: {len(self.filtrar_proyectos_por_estado('planificación'))}")
        print(f"  - Proyectos en Progreso: {len(self.filtrar_proyectos_por_estado('en_progreso'))}")
        print(f"  - Proyectos Completados: {len(self.filtrar_proyectos_por_estado('completado'))}")
        print("=================================================\n")
````

````python
from datetime import datetime
import uuid

# ==========================================
# BLOQUE INICIAL: HERRAMIENTAS GLOBALES
# ==========================================
def generar_id():
    return str(uuid.uuid4())[:8]

def validar_email(email):
    return "@" in email and "." in email

# ==========================================
# 1. REGISTRAR CLIENTES
# ==========================================
class Cliente:
    def __init__(self, nombre, email, sector):
        if not validar_email(email):
            raise ValueError("❌ Error: El formato del email del cliente no es válido.")
        self.id = generar_id()
        self.nombre = nombre
        self.email = email
        self.sector = sector
        self._historial_proyectos = []  # Protegido

    def registrar_proyecto_en_historial(self, nombre_proyecto):
        self._historial_proyectos.append(nombre_proyecto)

    def __str__(self):
        return f"Cliente: {self.nombre} | Sector: {self.sector} | ID: {self.id}"

    def __repr__(self):
        return f"Cliente('{self.nombre}', '{self.email}', '{self.sector}')"

# ==========================================
# 2. REGISTRAR DATASETS
# ==========================================
class Dataset:
    def __init__(self, nombre, fuente, filas, columnas, estado="crudo"):
        self.id = generar_id()
        self.nombre = nombre
        self.fuente = fuente
        self.fecha_registro = datetime.now().strftime("%Y-%m-%d %H:%M:%S")

        self._filas = None
        self._columnas = None
        self.filas = filas
        self.columnas = columnas

        estados_validos = ["crudo", "limpio", "procesado"]
        self.estado = estado if estado in estados_validos else "crudo"

    @property
    def filas(self):
        return self._filas

    @filas.setter
    def filas(self, valor):
        if valor < 0:
            raise ValueError("❌ Las filas no pueden ser negativas.")
        self._filas = valor

    @property
    def columnas(self):
        return self._columnas

    @columnas.setter
    def columnas(self, valor):
        if valor < 0:
            raise ValueError("❌ Las columnas no pueden ser negativas.")
        self._columnas = valor

    def calidad(self):
        if self.estado == "procesado": return 99.5
        if self.estado == "limpio": return 95.0
        return 70.0

    def __str__(self):
        return f"Dataset: {self.nombre} ({self.estado}) - {self.filas}x{self.columnas}"

# ==========================================
# 3. REGISTRAR ANALISTAS
# ==========================================
class Analista:
    MAX_PROYECTOS = 3

    def __init__(self, nombre, email, especialidad):
        if not validar_email(email):
            raise ValueError("❌ Error: El formato del email del analista no es válido.")
        self.nombre = nombre
        self.email = email
        self.especialidad = especialidad
        self.proyectos_activos = 0

    def disponible(self):
        return self.proyectos_activos < self.MAX_PROYECTOS

    def __str__(self):
        return f"Analista: {self.nombre} ({self.especialidad}) | Activos: {self.proyectos_activos}"

# ==========================================
# 4. CREAR PROYECTOS
# ==========================================
class Proyecto:
    def __init__(self, nombre, descripcion, cliente):
        if not isinstance(cliente, Cliente):
            raise TypeError("❌ El cliente debe ser un objeto válido de la clase Cliente.")

        self.nombre = nombre
        self.descripcion = descripcion
        self.cliente = cliente
        self.datasets = []
        self.equipo = []
        self._estado = "planificación"
        self._historial_estados = [f"Creado en 'planificación' el {datetime.now().strftime('%Y-%m-%d %H:%M:%S')}"]

        # Guardar en el historial del objeto cliente automáticamente
        self.cliente.registrar_proyecto_en_historial(self.nombre)

    @property
    def estado(self):
        return self._estado

    def agregar_dataset(self, dataset):
        if not isinstance(dataset, Dataset):
            raise TypeError("❌ Solo se pueden agregar objetos de tipo Dataset.")
        if dataset not in self.datasets:
            self.datasets.append(dataset)
            print(f"🔹 Dataset '{dataset.nombre}' vinculado al proyecto.")

    def asignar_analista(self, analista):
        if not isinstance(analista, Analista):
            raise TypeError("❌ Solo se pueden asignar objetos de tipo Analista.")

        if not analista.disponible():
            print(f"⚠️ No se pudo asignar: {analista.nombre} no tiene disponibilidad.")
            return False

        if analista not in self.equipo:
            self.equipo.append(analista)
            analista.proyectos_activos += 1
            print(f"🏃 Analista '{analista.nombre}' asignado al proyecto.")
            return True
        return False

    def cambiar_estado(self, nuevo_estado):
        estados_validos = ["planificación", "en_progreso", "completado"]
        if nuevo_estado not in estados_validos:
            raise ValueError(f"❌ Estado no válido. Debe ser uno de: {estados_validos}")

        if nuevo_estado != self._estado:
            registro = f"Cambio: {self._estado} -> {nuevo_estado} ({datetime.now().strftime('%Y-%m-%d %H:%M:%S')})"
            self._historial_estados.append(registro)
            self._estado = nuevo_estado
            print(f"🔄 Estado del proyecto actualizado a: {nuevo_estado.upper()}")

    def reporte(self):
        print("\n" + "="*50)
        print(f"📋 REPORTES DE PROYECTO: {self.nombre.upper()}")
        print(f"Estado: {self._estado.upper()}")
        print(f"Descripción: {self.descripcion}")
        print(f"Cliente: {self.cliente.nombre} | Sector: {self.cliente.sector}")
        print("-"*50)
        print(f"Equipo Técnico Asignado ({len(self.equipo)}):")
        for analista in self.equipo:
            print(f"  • {analista.nombre} ({analista.especialidad})")
        print(f"Datasets Utilizados ({len(self.datasets)}):")
        for ds in self.datasets:
            print(f"  • {ds.nombre} | Calidad: {ds.calidad()}%")
        print(f"Historial de Auditoría:")
        for cambio in self._historial_estados:
            print(f"  [Log] {cambio}")
        print("="*50 + "\n")

# ==========================================
# 5. BUSCAR, FILTRAR Y DATACOSYSTEM
# ==========================================
class DataCoSystem:
    def __init__(self):
        self.proyectos = []

    def registrar_proyecto(self, proyecto):
        if isinstance(proyecto, Proyecto):
            self.proyectos.append(proyecto)

    def filtrar_proyectos_por_estado(self, estado_buscado):
        return [p for p in self.proyectos if p.estado == estado_buscado]

    def generar_reporte_sistema(self):
        print("\n📊 === REPORTE GENERAL DE LA EMPRESA (DATACO) ===")
        print(f"Cantidad total de proyectos gestionados: {len(self.proyectos)}")
        print(f"  - Proyectos en Planificación: {len(self.filtrar_proyectos_por_estado('planificación'))}")
        print(f"  - Proyectos en Progreso: {len(self.filtrar_proyectos_por_estado('en_progreso'))}")
        print(f"  - Proyectos Completados: {len(self.filtrar_proyectos_por_estado('completado'))}")
        print("=================================================\n")


# ==========================================
# EJECUCIÓN DEL SISTEMA (EJEMPLO EN VIVO)
# ==========================================
print("🚀 Inicializando ejecuciones en limpio...")

sistema_dataco = DataCoSystem()

# 1. Registrar Cliente
cliente_finance = Cliente("Fintech S.A.", "ceo@fintech.com", "Finanzas")

# 2. Registrar Datasets
data_cruda = Dataset("Logs de Transacciones", "Servidor AWS", 80000, 15, "crudo")
data_procesada = Dataset("Muestra de Usuarios", "S3 Bucket", 2500, 8, "procesado")

# 3. Registrar Analista
analista_senior = Analista("Max Schell", "max.s@dataco.com", "Data Scientist")

# 4. Crear Proyecto y agrupar todo
proyecto_ia = Proyecto("Modelo Detección de Fraude", "Desarrollo de IA predictiva", cliente_finance)
proyecto_ia.agregar_dataset(data_cruda)
proyecto_ia.agregar_dataset(data_procesada)
proyecto_ia.asignar_analista(analista_senior)

sistema_dataco.registrar_proyecto(proyecto_ia)

# Actualizar el negocio
proyecto_ia.cambiar_estado("en_progreso")

# 5. Reportes
proyecto_ia.reporte()
sistema_dataco.generar_reporte_sistema()
````

### ¿Qué es `requirements.txt` y por qué es importante?

`requirements.txt` es un archivo de texto simple que contiene una lista de los paquetes Python que un proyecto necesita para funcionar. Es crucial para:

1.  **Reproducibilidad:** Asegura que cualquier persona que trabaje en el proyecto (o tú mismo en el futuro) pueda instalar exactamente las mismas versiones de las librerías, evitando problemas de compatibilidad.
2.  **Gestión de Dependencias:** Facilita la instalación de todas las dependencias del proyecto con un solo comando.
3.  **Entornos Virtuales:** Se usa comúnmente con entornos virtuales (`venv` o `conda`) para aislar las dependencias de un proyecto del resto de tu sistema Python.

````python
import os

# Añadiremos algunas librerías comunes de ciencia de datos a nuestro requirements.txt
# Puedes modificar esta lista según las necesidades reales de tu proyecto.

common_dependencies = [
    "pandas",
    "numpy",
    "scikit-learn",
    "matplotlib",
    "seaborn",
    "jupyter", # Para ejecutar notebooks, aunque ya estés en Colab
    "ipykernel" # Esencial para kernels de Python en entornos virtuales
]

requirements_dir = 'requirements'
# Ensure the directory exists before attempting to write the file
if not os.path.exists(requirements_dir):
    os.makedirs(requirements_dir)
    print(f"Directorio '{requirements_dir}' creado (desde esta celda).")

requirements_file_path = os.path.join(requirements_dir, 'requirements.txt')

with open(requirements_file_path, 'w') as f:
    for dep in common_dependencies:
        f.write(f"{dep}\n")
    print(f"Dependencies written to '{requirements_file_path}'.")

# Opcional: mostrar el contenido del archivo
print("\nContenido de requirements/requirements.txt:")
with open(requirements_file_path, 'r') as f:
    print(f.read())
````
